```python
import pandas as pd
import numpy as np

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
df = pd.read_csv("weather_madrid_LEMD_1997_2015.csv", low_memory=False)
```

```python
df.head()
```

```python
df.dtypes
```

```python
# change CET to date time format
df["CET"] = pd.to_datetime(df["CET"])
df = df.reset_index(drop=True)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
# Drop the columns we don't need
columns = ["Max Gust SpeedKm/h", "Precipitationmm", "Events", "WindDirDegrees"]
df = df.drop(columns = columns, axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = df.rename(columns={"CET": "Date"})
```

```python
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
df["Max Humidity"] = df["Max Humidity"]/100
df["Mean Humidity"] = df["Mean Humidity"]/100
df["Min Humidity"] = df["Min Humidity"]/100
```

```python
df = df.astype({"Max TemperatureC" : "int",
                "Mean TemperatureC": "int", 
                "Min TemperatureC": "int", 
                "Dew PointC": "int", 
                "MeanDew PointC": "int", 
                "Min DewpointC": "int", 
                "Max Humidity": "int", 
                "Mean Humidity": "int", 
                "Min Humidity": "int", 
                "Max Sea Level PressurehPa": "int", 
                "Mean Sea Level PressurehPa": "int", 
                "Min Sea Level PressurehPa": "int", 
                "Max VisibilityKm": "int", 
                "Mean VisibilityKm": "int",
                "Min VisibilitykM": "int", 
                "CloudCover": "int"})
```

```python
df.dtypes
```

```python
df.to_csv("Weather_Madrid_clean.csv", encoding='utf-8')
```
